```python
#### Basic Usage
import gradio as gr

# Define a function
def greet(name):
    return f"Hello, {name}!"

# Create a Gradio interface
demo = gr.Interface(
    fn=greet,  # Function to wrap
    inputs="text",  # Input component
    outputs="text"  # Output component
)

# Launch the interface
demo.launch()
```

```python
#### Input and Output Components

"""
Gradio supports various input and output components:
"""

import gradio as gr
import numpy as np
from PIL import Image

# Image processing function
def sepia(input_img):
    img_array = np.array(input_img)
    sepia_filter = np.array([
        [0.393, 0.769, 0.189],
        [0.349, 0.686, 0.168],
        [0.272, 0.534, 0.131]
    ])

    sepia_img = img_array.dot(sepia_filter.T)
    sepia_img /= sepia_img.max()
    sepia_img *= 255
    return sepia_img.astype(np.uint8)

# Create a Gradio interface with image input/output
demo = gr.Interface(
    fn=sepia,
    inputs=gr.Image(type="pil"),
    outputs=gr.Image(),
    title="Sepia Filter",
    description="Upload an image to apply a sepia filter."
)

demo.launch()
```

```python
#### Multiple Inputs and Outputs

import gradio as gr
import numpy as np
import matplotlib.pyplot as plt

def plot_function(function_type, x_min, x_max, num_points):
    x = np.linspace(x_min, x_max, num_points)

    if function_type == "sin":
        y = np.sin(x)
        title = "Sine Function"
    elif function_type == "cos":
        y = np.cos(x)
        title = "Cosine Function"
    elif function_type == "tan":
        y = np.tan(x)
        title = "Tangent Function"
    elif function_type == "exp":
        y = np.exp(x)
        title = "Exponential Function"
    else:
        y = x
        title = "Linear Function"

    fig, ax = plt.subplots()
    ax.plot(x, y)
    ax.set_title(title)
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.grid(True)

    return fig

demo = gr.Interface(
    fn=plot_function,
    inputs=[
        gr.Dropdown(["sin", "cos", "tan", "exp", "linear"], label="Function Type"),
        gr.Slider(-10, 10, 0, label="X Minimum"),
        gr.Slider(-10, 10, 5, label="X Maximum"),
        gr.Slider(10, 1000, 100, step=10, label="Number of Points")
    ],
    outputs="plot",
    title="Function Plotter"
)

demo.launch()
```

```python
#### Building a Chat Interface
import gradio as gr
import random

# List of potential responses
responses = [
    "That's interesting. Tell me more.",
    "I understand what you're saying.",
    "How does that make you feel?",
    "Let's explore that further.",
    "Can you elaborate on that?",
    "That's a good point."
]

def random_response(message, history):
    return random.choice(responses)

demo = gr.ChatInterface(
    fn=random_response,
    title="Simple Chatbot",
    description="This is a simple chatbot that responds randomly."
)

demo.launch()
```

```python
### 5.3 Deploying Gradio Apps to Hugging Face Spaces

"""
Hugging Face Spaces makes it easy to deploy Gradio applications for free.
"""

#### Steps to Deploy:

"""
1. Save your Gradio app in a file (e.g., `app.py`) --> It must be app.py for Hugging Face Spaces to recognize it as a Gradio app.
2. Create a `requirements.txt` file listing dependencies
3. Create a new Space on Hugging Face Spaces
4. Upload your files or connect to a GitHub repository
"""

#### Example `app.py`:

import gradio as gr

def calculator(operation, num1, num2):
    if operation == "add":
        return num1 + num2
    elif operation == "subtract":
        return num1 - num2
    elif operation == "multiply":
        return num1 * num2
    elif operation == "divide":
        if num2 == 0:
            return "Error: Division by zero"
        return num1 / num2

demo = gr.Interface(
    fn=calculator,
    inputs=[
        gr.Dropdown(["add", "subtract", "multiply", "divide"], label="Operation"),
        gr.Number(label="Number 1"),
        gr.Number(label="Number 2")
    ],
    outputs="text",
    title="Simple Calculator"
)

demo.launch()
```

```python
%%writefile simple_app.py
import streamlit as st

# Title and header
st.title("My First Streamlit App")
st.header("Welcome to Streamlit")

# Text input
name = st.text_input("Enter your name:")

# Button
if st.button("Greet"):
    st.write(f"Hello, {name}!")

# Display information
st.subheader("About this app")
st.write("This is a simple demonstration of Streamlit's capabilities.")

# Sidebar
st.sidebar.title("Navigation")
page = st.sidebar.radio("Go to", ["Home", "About", "Contact"])

if page == "Home":
    st.write("You are on the Home page")
elif page == "About":
    st.write("You are on the About page")
else:
    st.write("You are on the Contact page")

"""
# To run this app locally:
# 1. Save this code as simple_app.py
# 2. Run in terminal: streamlit run simple_app.py
"""
```

```python
### 5.5 Building a Number Guessing Game with Gradio

"""
Let's build a simple number guessing game using Gradio as our Project 3:
"""

import gradio as gr
import random

class NumberGuessingGame:
    def __init__(self):
        self.secret_number = random.randint(1, 100)
        self.attempts = 0
        self.max_attempts = 10
        self.game_over = False

    def reset_game(self):
        self.secret_number = random.randint(1, 100)
        self.attempts = 0
        self.game_over = False
        return "Game has been reset. I'm thinking of a new number between 1 and 100."

    def make_guess(self, guess):
        if self.game_over:
            return "The game is over. Click 'Reset Game' to play again."

        try:
            guess = int(guess)
        except ValueError:
            return "Please enter a valid number."

        self.attempts += 1

        if guess < 1 or guess > 100:
            return "Please enter a number between 1 and 100."

        if self.attempts >= self.max_attempts:
            self.game_over = True
            return f"Sorry, you've used all {self.max_attempts} attempts. The number was {self.secret_number}. Game over!"

        if guess < self.secret_number:
            return f"Too low! You have {self.max_attempts - self.attempts} attempts remaining."
        elif guess > self.secret_number:
            return f"Too high! You have {self.max_attempts - self.attempts} attempts remaining."
        else:
            self.game_over = True
            return f"Congratulations! You guessed the number in {self.attempts} attempts."

# Initialize the game
game = NumberGuessingGame()

# Define Gradio interface
with gr.Blocks(title="Number Guessing Game") as demo:
    gr.Markdown("# Number Guessing Game")
    gr.Markdown("I'm thinking of a number between 1 and 100. Can you guess it?")

    with gr.Row():
        with gr.Column(scale=3):
            guess_input = gr.Textbox(label="Enter your guess")
            guess_button = gr.Button("Make Guess")

        with gr.Column(scale=1):
            reset_button = gr.Button("Reset Game")

    output = gr.Textbox(label="Result", interactive=False)

    # Define event handlers
    guess_button.click(game.make_guess, inputs=guess_input, outputs=output)
    reset_button.click(game.reset_game, inputs=None, outputs=output)

    # Initial message
    gr.Markdown("You have 10 attempts to guess the number.")

# Launch the app
if __name__ == "__main__":
    demo.launch()
```
